**compton wavelength**

In this experiment, the Compton wavelength of the electron is determined by means of a direct method using X-rays. For this purpose, X-rays are scattered on a Plexiglas block. The energy of the scattered radiation as a function of the scattering angle is measured with a pivotable semiconductor detector and a multi-channel analyzer. A comparison with the known Compton formula then provides the Compton wavelength.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import NullFormatter
import matplotlib.ticker as ticker
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.preprocessing import label_binarize
from statistics import mean
from pprint import pprint
from scipy import optimize
from sklearn.preprocessing import label_binarize
from astropy.io import ascii
from astropy import modeling
```

```python
#X is the angle
#Y is the Energie in keV

#experimental values
X_1 = np.array([20, 48, 76, 104, 132, 160]).reshape((-1, 1))
Y_1 = np.array([8.07, 8.04, 8.03, 8.00, 7.97, 7.94])

#theoretical values
X_2 = np.array([20, 48, 76, 104, 132, 160]).reshape((-1, 1))
Y_2 = np.array([8.03, 8.00, 7.95, 7.89, 7.84, 7.80])
```

```python
plt.figure(figsize=(9, 9))
plt.xlabel("Winkel")
plt.ylabel("Enegrie (keV)")

#experimental curve
plt.plot(X_1, Y_1, color='blue', linestyle='dashed', linewidth = 1, marker='o', markerfacecolor='blue', markersize=8, 
         label='Experiment')

#theoretical curve
plt.plot(X_2, Y_2, color='red', linestyle='dashed', linewidth = 1, marker='o', markerfacecolor='red', markersize=8, 
         label='Theorie')

plt.legend(loc='best')
plt.show()
```

**Conclusion**

The Compton effect describes the change in the wavelength of photons when they scatter from electrically charged particles, e.g. Electrons, as a function of the scattering angle. The Compton effect can only be explained in the particle image of quantum theory, but not in the wave image.
