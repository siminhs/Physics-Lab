```python
import itertools
from itertools import cycle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import matplotlib.ticker as ticker
%matplotlib inline
from scipy import optimize
from scipy.optimize import curve_fit
from scipy import interp
from statistics import mean 
```

```python
#Ohne Linse
Position = [0, 0.5, 1, 1.5, 2, 2.25, 2.5, 2.75, 3, 3.25, 3.5, 3.75, 4, 4.25, 4.5, 4.75, 5, 5.5]
Leistung = [1708.249, 1703.025, 1687.4, 1656.369, 1585.081, 1447.209, 1295.044, 1024.144, 670.761, 345.744, 145.161, 50.176, 15.876, 7.569, 4.489, 4.096, 3.721, 3.6]

plt.plot(Position,Leistung)
plt.xlabel('Position (mm)')
plt.ylabel('Leistung (nW)')
```

```python
def func(z,a,b,c,d):
   return a*scipy.special.erf(b*(z-d))+c
```

```python
popt, pcov = curve_fit(func, Position, Leistung)
```

```python
popt
```

```python

x_data=Position
y_data=Leistung

def test_func(x, a, b, c, d):
    return a*scipy.special.erf(b*(x-d))+c

params, params_covariance = optimize.curve_fit(test_func, x_data, y_data,
                                               p0=[850, -20, 100, 0])

print(params)

plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, test_func(x_data, params[0], params[1],params[2],params[3]),
         label='Fitted function')

plt.legend(loc='best')

plt.show()
```
